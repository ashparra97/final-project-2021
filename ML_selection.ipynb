```python
import pandas as pd
```

```python
# csv_path = Data_Extraction_Transform/csv_misc/locationtype_month_earthquakes.csv
df = pd.read_csv('Data_Extraction_Transform/csv_misc/locationtype_month_earthquakes.csv', index_col=0)
df
```

```python
for index, row in df.iterrows():
    if df.loc[index, 'type'] == 'earthquake':
        df.loc[index,'grouped_type'] = 1
    else:
        df.loc[index,'grouped_type'] = 0
df.head(5)
```

```python
cleaned_df = df[['tsunami','felt', 'Unix Time', 'latitude', 'longitude', 'depth', 'mag', 'nst', 'gap', 'dmin', 'rms', 'grouped_type','magNst']]
cleaned_df
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

#https://medium.com/analytics-vidhya/how-to-find-the-best-performing-machine-learning-algorithm-dc4eb4ff34b6
```

### For Magnitude

```python
X = cleaned_df.drop(['mag','felt','nst','gap','dmin','magNst'], axis = 1)

y = cleaned_df[['mag']]
```

```python
X.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
scalar = StandardScaler()
```

```python
X_train_scaled = scalar.fit_transform(X_train)
```

```python
X_test_scaled = scalar.transform(X_test)
```

```python
# import
from lazypredict.Supervised import LazyClassifier
# initialize
clf = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None,predictions=True)
# run test
models,predictions = clf.fit(X_train_scaled, X_test_scaled, y_train, y_test)
```

```python
print(models)
print(predictions)
```

```python
# import
from lazypredict.Supervised import LazyRegressor

# initialize
reg = LazyRegressor(verbose=0,ignore_warnings=True, custom_metric=None)

# run test
models,predictions = reg.fit(X_train_scaled, X_test_scaled, y_train, y_test)

print(models)
print(predictions)
```

```python
from sklearn import neighbors
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error 
from math import sqrt

#https://www.analyticsvidhya.com/blog/2018/08/k-nearest-neighbor-introduction-regression-python/

rmse_val = [] #to store rmse values for different k
for K in range(20):
    K = K+1
    model = neighbors.KNeighborsRegressor(n_neighbors = K)

    model.fit(X_train, y_train)  #fit the model
    pred=model.predict(X_test) #make prediction on test set
    error = sqrt(mean_squared_error(y_test,pred)) #calculate rmse
    rmse_val.append(error) #store rmse values
    print('RMSE value for k= ' , K , 'is:', error)
```

```python
curve = pd.DataFrame(rmse_val) #elbow curve 
curve.plot()
```

Best Model

```python
from sklearn.model_selection import GridSearchCV
params = {'n_neighbors':[2,3,4,5,6,7,8,9]}

knn = neighbors.KNeighborsRegressor()

model = GridSearchCV(knn, params, cv=5)
model.fit(X_train,y_train)
model.best_params_
```

```python
K = 9
model = neighbors.KNeighborsRegressor(n_neighbors = K)

best_model = model.fit(X_train, y_train)  #fit the model
pred=best_model.predict(X_test) #make prediction on test set
error = sqrt(mean_squared_error(y_test,pred)) #calculate rmse
rmse_val.append(error) #store rmse values
print('RMSE value for k= ' , K , 'is:', error)
```

```python
predict = best_model.predict(X_test)
predicted = ([item[0] for item in predict])
actual = y_test['mag']
print(actual)
df_pred = pd.DataFrame({"Prediction Mag": predicted, "Actual Mag": actual})
df_pred
```

```python
pred_actual_clean_df = pd.merge(df_pred, cleaned_df, left_index=True, right_index=True)
pred_actual_clean_df
```
