## Download the csv from https://earthquake.usgs.gov/earthquakes/feed/v1.0/csv.php. Load it do pandas dataframe for necessary clean up if necessary (dropping columns not needed, changing date format, etc)

### Load csv from USGS

```python
csv_df = pd.read_csv('all_month(3).csv')
csv_df
```

### Create new dataframe from geojson to include tsunami and earthquake felt, in case it is needed for modeling

```python
small_df = df[['id', 'tsunami', 'felt']]
small_df
```

### Join csv dataframe and small geojson dataframe

```python
new_df = small_df.merge(csv_df, on="id", how = 'left')
new_df
```

```python
new_df.to_csv("month_earthquake.csv")
```

### Drop not needed columns

```python
clean_df = new_df.drop(['nst', 'dmin', 'rms', 'magNst', 'magError', 'depthError', 'horizontalError', 'magSource', 'status'], axis=1)
clean_df
```

```python
clean_df.to_csv("dropped_cols_month_earthquake.csv")
```

```python
felt_df = clean_df.dropna()
felt_df = felt_df.reset_index(drop=True)
felt_df
```

```python
felt_df.to_csv("dropped_NaN_month_earthquake.csv")
```
